# Results on the initial architecture
The architecture is composed of 11 Dense layers with 325 output neurons. The number of epochs is set to 25.
The testing set is roughly 20% of the entire dataset. The dataset is, in principle, extracted from HiggsML Data, but it falls back (due to errors yet misunderstood) to synthetic physical data.
Given this setup, the model achieves the following metrics:

**ROC AUC Score**: 0.76764

**Max Significance**: 3.9160
